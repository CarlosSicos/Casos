```python
import pandas as pd
from google.colab import files

# 1. Subir el archivo Excel
archivo = files.upload()

# 2. Leer el nombre del archivo subido
nombre_archivo = next(iter(archivo))

# 3. Cargar el archivo Excel (funciona con .xls o .xlsx)
df = pd.read_excel(nombre_archivo)

# 4. Guardar el archivo como CSV
nombre_csv = nombre_archivo.replace('.xls', '.csv').replace('.xlsx', '.csv')
df.to_csv(nombre_csv, index=False)

# 5. Descargar el archivo convertido
files.download(nombre_csv)
```

```python
archivo = files.upload()
df_academico = pd.read_csv(next(iter(archivo)))
df_academico.head()
```

```python
print(len(df_academico))
df_academico = df_academico.drop_duplicates()
print(len(df_academico))
```

```python
df_academico['Promedio' ] = pd.to_numeric(df_academico[ 'Promedio'], errors='coerce')
df_academico['Fecha_Ingreso' ] = pd.to_datetime(df_academico[ 'Fecha_Ingreso'], errors='coerce')
print(df_academico.info())
```
